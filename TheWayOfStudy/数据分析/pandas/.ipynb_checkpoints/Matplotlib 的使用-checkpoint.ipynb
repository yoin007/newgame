### About Sub-plots

```python
# plt a line, implicitly creating a subplot(111)
plt.plot([1, 2, 3])
# now create a subplot which represents the top plot of a grid
# with 2 rows and 1 column. Since this subplot will overlap the first,
# the plot(and its axes) previously created, will be removed
plt.subplot(211) #两行一列第一张图
plt.plot(range(15))
plt.subplot(212, facecolor = 'y') #第二张图creates 2nd subplot with yellow background
plt.plot(np.random.rand(10))
```

```python
plt.subplot(121) #一行两列第一张图
plt.plot(range(15))
plt.subplot(122, facecolor = 'y')
plt.plot(np.random.rand(10))
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)
#create some data to use for the plot
#*rect* = [left, bottom, width, height]

dt = 0.01
t = np.arange(0.0, 10.0, dt)
r = np.exp(-t[:1000]/0.05) # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt # colored noise
```

```python
# the main axes is subplot(111) by default
plt.plot(t, s)
plt.axis([0, 1, 1.1 * np.min(s), 2 * np.max(s)])
plt.xlabel('time(s)')
plt.ylabel('curent (nA)')
plt.title("Gaussian colored noise")

# this is an inset axes over the main axes
a = plt.axes([.65, .6, .2, .2], facecolor = 'k')
n, bins ,platches = plt.hist(s, 40, density =1)
plt.title('Probability')
plt.xticks([])
plt.yticks([])

# this is another inset axes over the main axes
a = plt.axes([0.2, 0.6, .2, .2], facecolor='k')
plt.plot(t[:len(r)],r)
plt.title('Impulse response')
plt.xlim(0, 0.2)
plt.xticks([])
plt.yticks([])

plt.show()
```

### 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = cm.viridis)
plt.show()
```

### What about Math Functions?

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.title(r'$\alpha_i > \beta_i$',fontsize = 20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$',fontsize = 20)
plt.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize = 20)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

```python
from scipy.misc import imread, imresize
```

```python
img = imread("./../../2.jpg")
img_tinted = img * [1, 1, 0.95]
print(img.shape)

img_tinted = imresize(img_tinted,(500,600))

# Show the original image
plt.subplot(1, 2, 1) #一行两列中的第一张图
plt.imshow(img)

# Show the tinted image
plt.subplot(1, 2, 2) #一行两列中的第二张图

# A siight gotcha with imshow is that it might give strange results
#if presented with data that is not uint8. To work around this, we
#explicitly cast the image to uint8 before displaying it.
plt.imshow(np.uint8(img_tinted))
plt.show()
```

#### Group by from dict and series

```python
data = DataFrame(np.random.randn(5,5), columns=['apple', 'orange', 'cucumber', 'banana', 'potato'], \
                 index = ['Monday', 'Tuesday', 'Wendesday', 'Thursday', 'Friday'])
data
```

```python
# That was an example of an aggregate, how about we make our own?
def range_values(arr):
    return arr.max() - arr.min()
```

```python
# apply our own aggregate function
#this function takes the max value of the col and subtracts the min value of the col
wino.agg(range_values)
```

```python
# We can also pass string methods through aggregate
wino.agg('sum')
```
